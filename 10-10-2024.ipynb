```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
num_rows = 30000
np.random.seed(42)  
data = {
    'BookingID': np.arange(1, num_rows + 1),
    'Airline': np.random.choice(['Air India', 'Vistara', 'Indigo', 'Spicejet'], num_rows),
    'FlightNumber': np.random.randint(1000, 9999, num_rows),
    'DepartureCity': np.random.choice(['Mumbai', 'Delhi', 'Kolkata', 'Guwahati','Chennai','Bangalore','Pune','Hyderabad'], num_rows),
    'ArrivalCity': np.random.choice(['Varanasi', 'Delhi', 'Kanpur', 'Patna','Vishakhapatnam','Kerela','Ladhak','Ayodhya','Indore','Ahmedabad','Patna','Sikkim'], num_rows),
    'DepartureTime': pd.date_range(start='2023-01-01', periods=num_rows, freq='H'),
    'ArrivalTime': pd.date_range(start='2023-01-01', periods=num_rows, freq='H') + pd.to_timedelta(np.random.randint(1, 10, num_rows), unit='h'),
    'PassengerCount': np.random.randint(1, 300, num_rows),
    'TicketPrice': np.round(np.random.uniform(50, 1000, num_rows), 2),
    'Class': np.random.choice(['Economy', 'Business', 'First'], num_rows),
    'BookingDate': pd.date_range(start='2022-01-01', periods=num_rows, freq='H'),
    'CancellationStatus': np.random.choice(['Cancelled', 'Not Cancelled'], num_rows),
    'LuggageWeight': np.round(np.random.uniform(0, 50, num_rows), 2)
}

df = pd.DataFrame(data)
```

```python
df
```

```python
# -------- Advanced Analytics --------

## 1. Feature Engineering
# Calculate travel duration
df['TravelDuration'] = (df['ArrivalTime'] - df['DepartureTime']).dt.total_seconds() / 3600  # Convert seconds to hours

# Calculate days between booking and departure
df['BookingLeadTime'] = (df['DepartureTime'] - df['BookingDate']).dt.days

# Binning ticket prices into categories
price_bins = [0, 200, 400, 600, 800, 1000]
price_labels = ['Very Low', 'Low', 'Medium', 'High', 'Very High']
df['PriceCategory'] = pd.cut(df['TicketPrice'], bins=price_bins, labels=price_labels)

# ## 2. Exploratory Data Analysis (EDA)
# # Correlation matrix
numeric_df = df.select_dtypes(include=[np.number])

# Create the correlation matrix heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# Distribution of Ticket Prices
plt.figure(figsize=(10, 6))
sns.histplot(df['TicketPrice'], bins=30, kde=True)
plt.title('Distribution of Ticket Prices')
plt.xlabel('Ticket Price')
plt.ylabel('Frequency')
plt.show()

# Average Ticket Price per Class
plt.figure(figsize=(10, 6))
sns.barplot(x='Class', y='TicketPrice', data=df, estimator=np.mean, ci=None)
plt.title('Average Ticket Price per Class')
plt.xlabel('Class')
plt.ylabel('Average Ticket Price')
plt.show()

# Cancellation Status by Class
plt.figure(figsize=(10, 6))
sns.countplot(x='Class', hue='CancellationStatus', data=df)
plt.title('Cancellation Status by Class')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()

# Passenger Count per Airline
plt.figure(figsize=(12, 6))
sns.boxplot(x='Airline', y='PassengerCount', data=df)
plt.title('Passenger Count per Airline')
plt.xlabel('Airline')
plt.ylabel('Passenger Count')
plt.show()

## 3. Advanced Analysis
# Average Travel Duration by Airline
avg_duration_by_airline = df.groupby('Airline')['TravelDuration'].mean()
print("Average Travel Duration by Airline:")
print(avg_duration_by_airline)

# Average Luggage Weight by Class and Price Category
avg_luggage_by_class_price = df.groupby(['Class', 'PriceCategory'])['LuggageWeight'].mean().unstack()
print("Average Luggage Weight by Class and Price Category:")
print(avg_luggage_by_class_price)

# Total Revenue per Airline
total_revenue_by_airline = df.groupby('Airline')['TicketPrice'].sum()
print("Total Revenue by Airline:")
print(total_revenue_by_airline)
```
